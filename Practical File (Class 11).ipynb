```python
#Q11)function. Prime or not prime.
def prime_notprime(n):
    sum = 0
    for i in range (1, n+1):
        if n%i==0:
            sum=sum+i
    if sum==1+n:
        return("1")
        print(n, "is prime")
    else:
        return("0")
        print(n, "is not prime")        
        
def print_prime_not(return_value, num_entered):
    if return_value=="1":
        print(num_entered,"is prime.")
    else:
        print(num_entered, "is not prime.")

num = int(input("Enter a number for checking "))
value=prime_notprime(num)
print(value)
print_prime_not(value,num)  
```
